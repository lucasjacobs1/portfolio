## Code of demo

```python
import random

def random_walk(n):
    """Return coordinates after 'n' block random walk"""
    x = 0
    y = 0
    for i in range(n):
        step = random.choice(['N', 'S', 'E', 'W'])
        if step == 'N':
            y = y + 1
        elif step == 'S':
            y = y - 1
        elif step == 'E':
            x = x + 1
        else:
            x = x - 1
    return (x, y)

for i in range(25):
    walk = random_walk(10)
    print(walk, "Distance from home = ",
          abs(walk[0]) + abs(walk[1]))



```

```python
def random_walk_2(n):
    """Return coordinates after 'n' block random walk."""
    x, y = 0, 0
    for i in range(n):
        (dx, dy) = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
        x += dx
        y += dy
    return (x, y)

for i in range(25):
    walk = random_walk_2(10)
    print(walk, "Distance from home = ",
        abs(walk[0]) + abs(walk[1]))
```

# Visuals of random walk

To begin with, we have the distribution plot of showing random walk coordinates. On the distribution we have every point representing a coordinate 
(x, y), reached during the random walks. The colour of the dots indicate the frequency of how many times the coordinate was visited. Side bar indicates the frequency values.

Next, the heatmap represent the frequency of visits to each coordinate, when making one iteration of the random walks. How coloured a certain block is, is based on frequency of visits to that block (coordinate). Lastly, the colour bar says something on what colour represents a certain frequency.

When we set the variable 'number_of_walks' to certain values (10k, 15k, 100k), the values will stay the same. It can always vary a bit because the monte carlo methods use random samples, so differenct simualations should give slithly different results. But when we look at the walk size 22, it will stay the same (around 50 procent), so it will always be the biggest walk size with the highest percentage of getting no transport. It is also possible to calculate confidence intervals for quantities such as the probability of no transport. With more simulations, the confidence interval will become smaller, which will indicate that there are more precise estimates. 


```python
import matplotlib.pyplot as plt
import numpy as np

def random_walk_3(n):
    """Return coordinates after 'n' block random walk."""
    x, y = 0, 0
    for i in range(n):
        (dx, dy) = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
        x += dx
        y += dy
    return (x, y)

number_of_walks = 10000

# Track coordinates and frequencies
coordinates = {}
for walk_length in range(1, 31):
    no_transport = 0  # Number of walks 4 or fewer blocks from home
    for i in range(number_of_walks):
        (x, y) = random_walk_3(walk_length)
        distance = abs(x) + abs(y)
        if distance <= 4:
            no_transport += 1
        if (x, y) in coordinates:
            coordinates[(x, y)] += 1
        else:
            coordinates[(x, y)] = 1
    no_transport_percentage = float(no_transport) / number_of_walks
    print("Walk size =", walk_length, "/ % of no transport =", 100 * no_transport_percentage)

# Create heatmap
heatmap_data = np.zeros((9, 9))
for coord, freq in coordinates.items():
    x, y = coord
    if abs(x) <= 4 and abs(y) <= 4:
        heatmap_data[x + 4][y + 4] = freq



plt.imshow(heatmap_data, cmap='hot', interpolation='nearest')
plt.title('Random Walk Heatmap')
plt.colorbar()
plt.show()

# Create distribution plot
x_coords = []
y_coords = []
freqs = []
for coord, freq in coordinates.items():
    x, y = coord
    x_coords.append(x)
    y_coords.append(y)
    freqs.append(freq)

plt.scatter(x_coords, y_coords, c=freqs, cmap='hot')
plt.title('Distribution of Random Walk Coordinates')
plt.colorbar(label='Frequency')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.show()
```

# Estimate the number Pi with Monte Carlo simulation

To visualize the results of pi, we use a scatter plot. when a random point is in the circle it will be added to the coordinates of the circle, otherwise, to the square. Basically, it generates random points in a square. When calculating pi, it counts the number of points in the circle that is in the square. After that, it will calculate an approximation of pi, based on the ratio of points in the circle. When it is done I visualized it with a scatter plot. The more samples I took, the better visualization it will become of a circle, so the estimate of pi is also more accurate.

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def estimate_pi(num_points):
    points_inside_circle = 0
    points_inside_square = 0

    x_coords_circle = []
    y_coords_circle = []
    x_coords_square = []
    y_coords_square = []

    for _ in range(num_points):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        distance = x**2 + y**2

        if distance <= 1:
            points_inside_circle += 1
            x_coords_circle.append(x)
            y_coords_circle.append(y)
        else:
            x_coords_square.append(x)
            y_coords_square.append(y)

        points_inside_square += 1
    #Calculates pi
    pi_estimate = 4 * points_inside_circle / points_inside_square
    return pi_estimate, x_coords_circle, y_coords_circle, x_coords_square, y_coords_square

#amount of points to take estimate of pi
num_points = 1000000

pi_estimate, x_circle, y_circle, x_square, y_square = estimate_pi(num_points)
print("Estimated value of pi:", pi_estimate)

# Create a scatter plot to visualize the square and circle
plt.scatter(x_square, y_square, color='blue', label='Outside Circle')
plt.scatter(x_circle, y_circle, color='red', label='Inside Circle')

#aspect ratio to equal, so the square appears as a square
plt.gca().set_aspect('equal')

# Add circle with radius 1
circle = plt.Circle((0, 0), 1, edgecolor='black', facecolor='none')
plt.gca().add_patch(circle)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.title('Monte Carlo Estimation of Pi')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.legend()
plt.show()


```
